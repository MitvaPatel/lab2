```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/MitvaPatel/Datasets/main/data.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Select the features by column names
features = df.loc[:, 'radius_mean':'fractal_dimension_worst']

# Select the target variable
target = df['diagnosis']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Create an instance of SVC
clf = SVC()
```

```python
# Fit the classifier on the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = clf.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)*100
print(f"SVM Accuracy: {accuracy:.2f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create a heatmap of the confusion matrix
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d')

# Add labels and title
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Convert labels to binary format (0 for 'B', 1 for 'M')
label_encoder = LabelEncoder()
y_test_binary = label_encoder.fit_transform(y_test)

# Calculate the decision scores
decision_scores = clf.decision_function(X_test)

# Manually calculate the probabilities
probs = (decision_scores - decision_scores.min()) / (decision_scores.max() - decision_scores.min())

# Vary the threshold and calculate accuracy for each threshold
thresholds = np.linspace(0, 1, num=100)
accuracies = []

for threshold in thresholds:
    # Classify samples based on the threshold
    y_pred = (probs >= threshold).astype(int)
    accuracy = accuracy_score(y_test_binary, y_pred)
    accuracies.append(accuracy)

# Plot the accuracy graph
plt.plot(thresholds, accuracies)
plt.xlabel('Threshold')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Threshold')

# Display the chart
plt.show()
```
